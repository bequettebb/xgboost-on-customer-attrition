```python
import numpy as np
from sklearn.ensemble import IsolationForest

np.random.seed(42)
X_train = 0.2 * np.random.randn(1000, 6)
X_train = np.r_[X_train + 3, X_train]
# Add some outliers
X_outliers = np.random.uniform(low=-1, high=5, size=(50, 6))
dfFF = pd.DataFrame(np.r_[X_train, X_outliers], columns=['Total_Purchases', 'Average_Order_Value','Days_Since_Last_Purchase','Discount_Usage_Rate','Returns_Rate','Payment_Method_Diversity'])
```

```python
# Define the features to use for outlier detection
anomaly_inputs = ['Total_Purchases', 'Average_Order_Value','Days_Since_Last_Purchase','Discount_Usage_Rate','Returns_Rate','Payment_Method_Diversity']


# Initialize the model (set contamination based on your domain knowledge)
model_IF = IsolationForest(contamination=0.05, random_state=42) # Assuming 5% outliers

# Fit the model to your data
model_IF.fit(dfFF[anomaly_inputs])
```

```python
# Get anomaly scores and add them to a new column
dfFF['anomaly_scores'] = model_IF.decision_function(dfFF[anomaly_inputs])

# Get the binary predictions and add them to a new column
dfFF['anomaly_label'] = model_IF.predict(dfFF[anomaly_inputs])

# View the DataFrame with the new columns
print(dfFF.head())
print(dfFF['anomaly_label'].value_counts()) # Count the number of inliers (1) and outliers (-1)
```

```python
# Filter for normal data points (label == 1)
df_inliers = dfFF[dfFF['anomaly_label'] == 1]

# Filter for outliers (label == -1)
df_outliers = dfFF[dfFF['anomaly_label'] == -1]

print("\nDataFrame without outliers:")
print(df_inliers.head())
```
